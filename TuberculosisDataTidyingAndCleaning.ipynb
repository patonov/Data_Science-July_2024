```python
def process_age_group(age_group):
    ages = {"04": "0-4", "65": "65+", "u": "unknown"}
    if age_group in ages:
        return ages[age_group]

    return f"{age_group[:-2]}-{age_group[-2:]}"
```

```python
df_tidy.age_group = df_tidy.age_group.apply(process_age_group)
```

```python
df_tidy.age_group.unique()
```

```python
df_tidy
```

```python
df_tidy.gender = df_tidy.gender.astype("category")
```

```python
df_tidy.age_group = df_tidy.age_group.astype("category")
```

```python
df_tidy = df_tidy[["iso2", "year", "age_group", "gender", "cases"]]
```

```python
df_tidy
```

```python
df_tidy = df_tidy.sort_values(["iso2", "year"])
```

```python
df_tidy = df_tidy.reset_index(drop=True)
```

```python
df_tidy.cases = df_tidy.cases.astype("int")
```

```python
df_tidy.to_csv("./input/tb_tidy.csv", index=None)
```
